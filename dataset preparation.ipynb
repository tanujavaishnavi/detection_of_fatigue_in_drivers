```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
raw_dir = r'C:\Users\panja\Downloads\mrlEyes_2018_01'
for dirpath,dirname,filenames in os.walk(raw_dir):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4] == '0':
            shutil.copy(src=dirpath+'/'+i,dst=r'C:\Users\panja\Downloads\eyeClose')
        elif i.split('_')[4] == '1':
            shutil.copy(src=dirpath+'/'+i,dst=r'C:\Users\panja\Downloads\eyeOpen')
```
